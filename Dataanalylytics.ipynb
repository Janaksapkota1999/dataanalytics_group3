```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
df = pd.read_csv('downloads/Data_Manipulation/Employment Statistics.csv')

# Check the data
plt.figure(figsize=(12, 6))
sns.boxplot(x='Median employment income ($)', y='Employment income statistics', data=df)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Median Employment Income ($)')
plt.ylabel('Major Field of Study')
plt.title('Median Employment Income by Major Field of Study')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
df = pd.read_csv('downloads/Data_Manipulation/Employment Statistics.csv')
plt.figure(figsize=(12, 6))
sns.barplot(x='Total - Employment income statistics', y='Employment income statistics', data=df)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Major Field of Study')
plt.ylabel('Total Number of People')
plt.title('Total Employment Income Statistics by Major Field of Study')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(12, 6))
sns.violinplot(x='Average employment income ($)', y='Employment income statistics', data=df)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Average Employment Income ($)')
plt.ylabel('Major Field of Study')
plt.title('Average Employment Income by Major Field of Study')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('downloads/Data_Manipulation/Employment Statistics.csv')
plt.figure(figsize=(8, 8))
sns.scatterplot(x='Median employment income ($)', y='Average employment income ($)', data=df)
plt.xlabel('Median Employment Income ($)')
plt.ylabel('Average Employment Income ($)')
plt.title('Comparison of Median and Average Employment Income')
plt.show()


```
